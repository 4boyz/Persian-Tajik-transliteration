```python
from tj_pers_checker import TjPersChecker   # type: ignore
from bbc_parser import BBCParser            # type: ignore
from dump_reader import DumpReader          # type: ignore

import pandas as pd
```

```python
PATH_TO_MATCHED_ARTICLES = '../data/matched_links_full.csv'

PATH_TO_PERSIAN_DUMP = 'C:\\Users\\mirvo\\Downloads\\Telegram Desktop\\persian_dump\\persian_dump'
PATH_TO_TJIK_DUMP = 'C:\\Users\\mirvo\\Downloads\\Telegram Desktop\\tajik_dump\\tajik_dump'

PATH_TO_SAVE = '../data/sen_to_sen.csv'
```

```python
matched_links = pd.read_csv(PATH_TO_MATCHED_ARTICLES)
matched_links
```

```python
matched_links = matched_links.drop(columns=['Unnamed: 0'])
matched_links = matched_links.reset_index(drop=True)
matched_links = matched_links.sort_values(by=['prob'])
matched_links['id'] = matched_links['id'].astype(int) 
matched_links = matched_links[matched_links['prob'] > .63]
matched_links
```

```python
persian_ids = matched_links['id'].to_list()
tajik_ids = matched_links['tajik_id'].to_list()
```

```python
set2sen = []
```

```python
for pid, tid in zip(persian_ids, tajik_ids):
  set2sen.extend(TjPersChecker.match_sentences(BBCParser.parse_text_tj(DumpReader.read_item_by_id(PATH_TO_TJIK_DUMP, tid)['Content']), BBCParser.parse_text_pers(DumpReader.read_item_by_id(PATH_TO_PERSIAN_DUMP, pid)['Content'])))
```

```python
pd.DataFrame(set2sen).to_csv('../data/sentence_to_sentence.csv')
```
