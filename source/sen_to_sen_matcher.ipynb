```python
from tj_pers_checker import TjPersChecker   # type: ignore
from bbc_parser import BBCParser            # type: ignore
from dump_reader import DumpReader          # type: ignore

import pandas as pd
```

```python
PATH_TO_MATCHED_ARTICLES = '../data/matched_links_full.csv'

PATH_TO_PERSIAN_DUMP = 'C:\\Users\\mirvo\\Downloads\\Telegram Desktop\\persian_dump\\persian_dump'
PATH_TO_TJIK_DUMP = 'C:\\Users\\mirvo\\Downloads\\Telegram Desktop\\tajik_dump\\tajik_dump'

PATH_TO_SAVE = '../data/sen_to_sen.csv'
```

```python
matched_links = pd.read_csv(PATH_TO_MATCHED_ARTICLES)
matched_links
```

```python
matched_links = matched_links.sort_values(by=['prob'])
matched_links = matched_links[matched_links['prob'] > .63]
matched_links['pair_id'] = matched_links.index
matched_links['tj_id'] = matched_links['tajik_id'].astype(int)
matched_links['tj_title'] = matched_links['tajik_title']
matched_links['tj_link'] = matched_links['tajik_link']
matched_links['fa_id'] = matched_links['id'].astype(int)
matched_links['fa_title'] = matched_links['title']
matched_links['fa_link'] = matched_links['link']

matched_links = matched_links.drop(columns=['tajik_id', 'tajik_title', 'tajik_link', 'id', 'title', 'link', 'Unnamed: 0'])
matched_links
```

```python
matched_links_dict = matched_links.to_dict('records')
matched_links_dict
```

```python
set2sen = []
```

```python
for mathched_pair in matched_links_dict:
  tid = mathched_pair['tj_id']
  pid = mathched_pair['fa_id']
  set2sen.extend(
    TjPersChecker.match_sentences(
      BBCParser.parse_text_tj(DumpReader.read_item_by_id(PATH_TO_TJIK_DUMP, tid)['Content']), 
      BBCParser.parse_text_pers(DumpReader.read_item_by_id(PATH_TO_PERSIAN_DUMP, pid)['Content']),
      other_info=mathched_pair),
      )
```

```python
set2sen
```

```python
pd.DataFrame(set2sen).to_csv('../data/sentence_to_sentence.csv', encoding='utf-8-sig', index=False)
```
