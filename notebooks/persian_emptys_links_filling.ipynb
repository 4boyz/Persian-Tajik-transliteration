```python
from pandas import Series, isna


def get_beauty_link(link):
    if not link: return ''
    try:
        return requests.get(link).url
    except:
        pass
    return ''

def get_last_link(content: str):
    return content.split(' ')[-1]

# ('id',          'Date'),
# ('id',          'User'),
# ('id',         'Links'),
# ('id',       'Content'),
# ('id', 'LinkFormatted'),
# ('id',          'Some')

def try_get_link(row: Series):
    links = row['Links']
    tweet_content = row['Content']
    first_links = ast.literal_eval(links if links and not links else '[]')
    beauty_link = ''
    if first_links:
        beauty_link = get_beauty_link(first_links[-1])
    else:
        beauty_link = get_beauty_link(get_last_link(tweet_content))
    row['LinkFormatted'] = beauty_link
    print(row)
    return row

```
